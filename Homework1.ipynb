#### 1. Посмотреть документацию к API GitHub, разобраться как вывести список репозиториев для конкретного пользователя, сохранить JSON-вывод в файле *.json.

```python
import requests
resp = requests.get("https://api.github.com/users/AlexanderMaslikhin/repos")
```

```python
repo_lst = resp.json()
for repo in repo_lst:
  print(f'{repo["name"]}\n\towner: {repo["owner"]["login"]}\n\turl: {repo["html_url"]}\n\tlanguage: {repo["language"]}\n\tdescription: {repo["description"]}')
```
